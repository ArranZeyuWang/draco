# Count violations

Show the violations of a spec. 

```python
from draco.spec import Data
from draco.util import count_violations
import pandas as pd
from vega3 import VegaLite
```

```python
data_url = '../examples/data/cars.csv'
data = Data.from_csv(data_url)

spec = {
    "encoding": {
        "x": {
            "bin": 10,
            "field": "horsepower",
            "type": "quantitative"
        },
        "y": {
            "aggregate": "count",
            "scale": {
                "zero": True
            },
            "type": "quantitative"
        }
    },
    "mark": "bar"
}
```

```python
VegaLite(spec, pd.read_csv(data_url))
```

```python
violations = count_violations(spec, data)
violations
```
